```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image, ImageOps
from cv2 import cv2
```

```python
# plate
row = 8
columns = 12
```

```python
path = 'images/plate.jpg'

image = cv2.imread(path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
blurred = cv2.medianBlur(gray, 3)
```

```python
height, width, channels = image.shape 

diameter = ((height/row) + (width/columns)) / 2
radius = diameter / 2
print(radius)
```

```python
param1 = 30 #30 #500
param2 = 20 #50 #200 #smaller value-> more false circles
minRadius = int(radius - radius*0.25) # 15
maxRadius =  int(radius + radius*0.15) # 25
print(minRadius, maxRadius)
minDist = (minRadius+maxRadius)/2
canny = 1
```

```python
circles = cv2.HoughCircles(blurred, cv2.HOUGH_GRADIENT, canny, minDist, param1=param1, param2=param2, minRadius=minRadius, maxRadius=maxRadius)
```

```python
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        cv2.circle(image, (i[0], i[1]), i[2], (0, 255, 0), 2)
```
